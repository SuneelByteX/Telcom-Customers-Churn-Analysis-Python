```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('Customer Churn.csv')
df.head()
```

```python
df.info()
```

#replacing blacks with 0 as tenure is 0 and no total charges are recorded

```python
df["TotalCharges"] = df["TotalCharges"].replace(" ","0")
df["TotalCharges"] = df["TotalCharges"].astype("float")
```

```python
df.info()
```

```python
df.isnull().sum().sum()
```

```python
df.describe()
```

```python
df["customerID"].duplicated().sum()
```

```python
def conv(value):
    if value == 1:
        return "yes"
    else:
        return "no"

df["SeniorCitizen"] = df["SeniorCitizen"].apply(conv)
```

#converted 0 and 1 values of senior citizen to yes/no to make it easier to understand

```python
df.head(30)
```

```python
ax = sns.countplot(x = "Churn", data = df)

ax.bar_label(ax.containers[0])
plt.title("Count of Customer by Churn")
plt.show()  # Display the plot
```

```python
plt.figure(figsize = (3,4))
gb = df.groupby("Churn").agg({"Churn":"count"})
plt.pie(gb['Churn'], labels = gb.index, autopct = "%1.2f%%")
plt.title("Percentage of Churned Customeres", fontsize = 10)
plt.show()  # Display the plot
```

#from the given pie chart we can conclude that 26.54% of our customer have churned out. 
#Know let's explore the reason behind it

```python
plt.figure(figsize = (3,3))
sns.countplot(x = "gender", data = df, hue = "Churn")
plt.title("Churn by Gender")
plt.show()  # Display the plot
```

```python
plt.figure(figsize = (4,4))
ax = sns.countplot(x = "SeniorCitizen", data = df)
ax.bar_label(ax.containers[0])
plt.title("Churn of Customers by  Senior Citizen")
plt.show()  # Display the plot
```

```python
churn_counts = pd.crosstab(df["SeniorCitizen"], df["Churn"], normalize="index") * 100

# Plot
plt.figure(figsize=(4,4))
churn_counts.plot(kind="bar", stacked=True, colormap="viridis", edgecolor="black")

# Add labels
for index, row in enumerate(churn_counts.values):
    plt.text(index, row[0] / 2, f"{row[0]:.1f}%", ha="center", color="white", fontsize=12, fontweight="bold")
    plt.text(index, row[0] + row[1] / 2, f"{row[1]:.1f}%", ha="center", color="white", fontsize=12, fontweight="bold")

# Title and labels
plt.title("Churn by SeniorCitizen (Stacked %)", fontsize=14)
plt.xlabel("Senior Citizen", fontsize=12)
plt.ylabel("Percentage", fontsize=12)
plt.legend(title="Churn", labels=["No", "Yes"])
plt.xticks(rotation=0)
plt.show()  # Display the plot
```

#comparative a greater percentage of people in senior citizen category have churned

```python
plt.figure(figsize = (9,4))
sns.histplot(x = "tenure", data = df, bins = 72, hue = "Churn")
plt.show()  # Display the plot
```

```python
# Define the columns to plot
columns = ['PhoneService', 'MultipleLines', 'InternetService',
           'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
           'TechSupport', 'StreamingTV', 'StreamingMovies']

# Define subplot grid
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(15, 12))  # 3x3 grid

# Flatten axes array for easy iteration
axes = axes.flatten()

# Loop through columns and create countplots
for i, col in enumerate(columns):
    sns.countplot(x=df[col], hue=df["Churn"], data=df, ax=axes[i], palette='viridis', "viridis", 
                  edgecolor="black")  # hue set to Churn
    axes[i].set_title(f"Churn Distribution by {col}")
    axes[i].set_xlabel("")
    axes[i].set_ylabel("Count")
    axes[i].tick_params(axis='x', rotation=45)

# Adjust layout
plt.tight_layout()
plt.show()  # Display the plot
```

#The majority of customers who do not churn tend to have services like PhoneService, Internetservice (particularly DSL).and OnlineSecurity enabled for
onlinebackup. TechSupport, and StreamingTV, churn rates are noticeably higher when these services are not used or are unavailable.
